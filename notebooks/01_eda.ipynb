# Exploratory Data Analysis - NASA CMAPSS Dataset
This notebook explores the turbofan engine sensor data to understand degradation patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add src to path
sys.path.append('../src')
from data_preprocessing import load_data, calculate_rul

%matplotlib inline
sns.set_style('whitegrid')
```

```python
# Load FD001 Train Data
train_df = load_data('../data/train_FD001.txt')
train_df = calculate_rul(train_df)
train_df.head()
```

## Sensor Visualization
Let's look at how specific sensors change over time for a single engine unit.

```python
unit_id = 1
unit_data = train_df[train_df['unit_nr'] == unit_id]

sensors_to_plot = ['s_2', 's_3', 's_4', 's_7', 's_11', 's_12']

plt.figure(figsize=(15, 10))
for i, sensor in enumerate(sensors_to_plot, 1):
    plt.subplot(3, 2, i)
    plt.plot(unit_data['time_cycles'], unit_data[sensor])
    plt.title(f'Sensor {sensor} - Unit {unit_id}')
    plt.xlabel('Cycles')
    plt.ylabel('Reading')

plt.tight_layout()
plt.show()
```

## RUL Distribution
How many cycles do engines typically run before failure?

```python
max_cycles = train_df.groupby('unit_nr')['time_cycles'].max()
plt.figure(figsize=(10, 6))
sns.histplot(max_cycles, bins=20, kde=True, color='teal')
plt.title('Distribution of Max Cycles (Lifetime) per Engine')
plt.xlabel('Cycles')
plt.show()
```
